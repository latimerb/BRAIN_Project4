```python
import sys
%matplotlib inline  

cell_file = '../cells/olm.cell.nml'
cell_id = 'olmcell'

from pyneuroml.analysis import generate_current_vs_frequency_curve
    
    
curve = generate_current_vs_frequency_curve(cell_file, 
                                    cell_id, 
                                    custom_amps_nA =       [-0.1, -0.07, -0.04, -0.01, 0.23], 
                                    analysis_duration =    1000, 
                                    pre_zero_pulse =       50,
                                    post_zero_pulse =      50,
                                    analysis_delay =       0,
                                    dt =                   0.025,
                                    simulator =            'jNeuroML_NEURON',
                                    plot_voltage_traces =  True,
                                    plot_if =              False,
                                    plot_iv =              False,
                                    temperature =          '34degC',
                                    title_above_plot =      True)
```
